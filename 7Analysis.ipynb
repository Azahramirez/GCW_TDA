# ANOVA

```python
import numpy as np

# load a numpy array
data = np.load('data/acc_completo.npy')
data.shape
# 10 Rs, 8 models, 10 folds
```

```python
data[0]
```

we need to extract the last model results of each R

```python
resultados_tda= data[:, 7, :]  # TDA CNN results across Rs and folds
print(resultados_tda.shape)  # should be (10, 10) for 10 Rs
resultados_tda[0]
```

Now only the rand(n) will remain for the ANOVA analysis

```python
resultados_tda_rand= resultados_tda[7:10, :]  # TDA CNN results for rand(0.065)
print(resultados_tda_rand.shape)  # should be (3, 10) for 10 Rs
resultados_tda_rand
```

```python
from scipy.stats import f_oneway
from scipy.stats import shapiro

grupo1 = resultados_tda_rand[0, :]  # rand100
grupo2 = resultados_tda_rand[1, :]  # rand250
grupo3 = resultados_tda_rand[2, :]  # rand500
```

```python

for i, grupo in enumerate([grupo1, grupo2, grupo3], start=1):
    stat, p = shapiro(grupo)
    print(f"Grupo {i}: p = {p:.4f} {' Normal' if p > 0.05 else ' No normal'}")
```

Log transformation

```python
from scipy.stats import boxcox

grupo1_boxcox, lab1 = boxcox(grupo1 )  # Adding 1 to avoid log(0)
print(f"Grupo 1 Box-Cox: {lab1}...")  # Display first 5 values for brevity

grupo2_boxcox, lab2 = boxcox(grupo2 )  # Adding 1 to avoid log(0)
print(f"Grupo 2 Box-Cox: {lab2}...")  # Display first 5 values for brevity

grupo3_boxcox, lab3 = boxcox(grupo3 )  # Adding 1 to avoid log(0)
print(f"Grupo 3 Box-Cox: {lab3}...")  # Display
```

```python
for i, grupo in enumerate([grupo1_boxcox, grupo2_boxcox, grupo3_boxcox], start=1):
    stat, p = shapiro(grupo)
    print(f"Grupo {i}: p = {p:.4f} {' Normal' if p > 0.05 else ' No normal'}")
```

```python

f_stat, p_valor = f_oneway(grupo1, grupo2, grupo3)

print("Estadstico F:", f_stat)
print("Valor p:", p_valor)
```

```python
from scipy.stats import bartlett

stat, p = bartlett(grupo1, grupo2, grupo3)
print(f"Bartlett test: p = {p:.4f} {' Varianzas iguales' if p > 0.05 else ' Varianzas distintas'}")
```
